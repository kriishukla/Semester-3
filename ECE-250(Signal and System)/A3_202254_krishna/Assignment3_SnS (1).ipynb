```python
import numpy as np
import matplotlib.pyplot as plt

def periodic_x(n, N1, N):
    if N1 < abs(n) <= (N - 1) / 2:
        return 0
    elif abs(n) <= N1:
        return 1
    elif N1 <= n <= N - 1:
        return periodic_x(n - N, N1, N)
    elif -N + 1 <= n <= -N1:
        return periodic_x(n + N, N1, N)
    return 0

def y(N, N1):
    N_values = np.arange(-N, N + 1)
    coefficients = []

    for k in range(-N, N + 1):
        real_sum = sum(periodic_x(n, N1, N) * np.cos(2 * np.pi * k * n / (2 * N1 + 1)) for n in N_values)
        imaginary_sum = sum(periodic_x(n, N1, N) * np.sin(2 * np.pi * k * n / (2 * N1 + 1)) for n in N_values)
        coefficient = (1 / (2 * N1 + 1)) * (real_sum - 1j * imaginary_sum)
        coefficients.append(coefficient)

    return coefficients

def plot_fourier_series(N, N1):
    coefficients = y(N, N1)
    N_values = np.arange(-N, N + 1)

    # Define the frequency and angular frequency (w0)
    w0 = 2 * np.pi / (2 * N1 + 1)

    # Create arrays to store the real and imaginary parts of the Fourier series
    real_part = np.zeros(len(N_values))
    imag_part = np.zeros(len(N_values))

    for k in range(-N, N + 1):
        a_k = coefficients[k + N]
        for n in range(len(N_values)):
            real_part[n] += np.real(a_k * np.exp(-1j * k * w0 * N_values[n]))
            imag_part[n] += np.imag(a_k * np.exp(-1j * k * w0 * N_values[n]))

    # Plot the real and imaginary parts of the Fourier series
    plt.figure(figsize=(10, 4))

    plt.subplot(2, 1, 1)
    plt.stem(N_values, real_part, 'b', basefmt='b-')
    plt.title("Real part of Fourier Series")
    plt.xlabel('n')
    plt.ylabel('f(n)')
    plt.grid()

    plt.subplot(2, 1, 2)
    plt.stem(N_values, imag_part, 'g', basefmt='g-')
    plt.title("Imaginary part of Fourier Series")
    plt.xlabel('n')
    plt.ylabel('f(n)')
    plt.grid()

    plt.tight_layout()
    plt.show()

# (a) N = 4N1 + 1
N_a = 4 * 2 + 1
N1_a = 2
plot_fourier_series(N_a, N1_a)

# (b) N = 8N1 + 1
N_b = 8 * 2 + 1
N1_b = 2
plot_fourier_series(N_b, N1_b)

# (c) N = 10N1 + 1
N_c = 10 * 2 + 1
N1_c = 2
plot_fourier_series(N_c, N1_c)
```
