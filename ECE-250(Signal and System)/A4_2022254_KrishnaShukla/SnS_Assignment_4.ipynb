```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function for the summation
def summation_function(x):
    return np.sum(x - np.pi * np.arange(-1000, 1001))

# Generate x values
x_values = np.linspace(-10, 10, 1000)

# Calculate the y values for the summation function
y_values = np.vectorize(summation_function)(x_values)

# Plot the summation function
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r'$\sum_{k=-\infty}^{\infty} (w - \pi k)$')
plt.xlabel('w')
plt.ylabel('Summation')
plt.title('Summation Function')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the piecewise function for x(t)
def piecewise_function(t):
    if -1 <= t < 0:
        return 2
    elif 0 <= t < 1:
        return -t + 2
    elif 1 <= t < 2:
        return t
    elif 2 <= t < 3:
        return 2
    else:
        return 0

def X_jomega(omega):
    integral_real = lambda t: piecewise_function(t) * np.cos(omega * t)
    integral_imag = lambda t: piecewise_function(t) * np.sin(omega * t)

    real_part, _ = quad(integral_real, -np.inf, np.inf)
    imag_part, _ = quad(integral_imag, -np.inf, np.inf)

    return real_part + 1j * imag_part

omega_values = np.linspace(-5, 5, 1000)
complex_result = [X_jomega(omega) for omega in omega_values]

plt.figure(figsize=(10, 6))
plt.plot(omega_values, np.real(complex_result), label='Real Part')
plt.plot(omega_values, np.imag(complex_result), label='Imaginary Part')
plt.xlabel('Frequency ()')
plt.ylabel('X(j)')
plt.title('Real and Imaginary Parts of X(j)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def piecewise_function(t):
    if -1 <= t < 0:
        return 2
    elif 0 <= t < 1:
        return -t + 2
    elif 1 <= t < 2:
        return t
    elif 2 <= t < 3:
        return 2
    else:
        return 0

def X_jomega(omega):
    integral = lambda t: piecewise_function(t) * np.exp(-1j * omega * t)
    result, _ = quad(integral, -np.inf, np.inf)
    return result

omega_values = np.linspace(-5, 5, 1000)

magnitude_values = np.abs([X_jomega(omega) for omega in omega_values])

plt.figure(figsize=(10, 6))
plt.plot(omega_values, magnitude_values, label='Magnitude')
plt.xlabel('Frequency ()')
plt.ylabel('|X(j)|')
plt.title('Magnitude of X(j)')
plt.legend()
plt.grid(True)
plt.show()
```
